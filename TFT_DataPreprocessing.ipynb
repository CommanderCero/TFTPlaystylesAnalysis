```python
import pandas as pd
import json
```

```python
# combination_set = set()

# for combination in df.combination:
#     for i in combination:
#         combination_set.add(i)

# combination_index_dict = {combination: i for i, combination in enumerate(combination_set, start=0)}
# json.dump(combination_index_dict, open("combination_index.json", "w"), indent=4)

with open("combination_index.json") as f:
    combination_index_dict = json.load(f)
    
combination_index_dict
```

```python
# champion_set = set()

# for champion in df.champion:
#     for i in champion:
#         champion_set.add(i)
        
# champion_index_dict = {champion: i for i, champion in enumerate(champion_set, start=1)}
# json.dump(champion_index_dict, open("champion_index.json", "w"), indent=4)

with open("champion_index.json") as f:
    champion_index_dict = json.load(f)
    
champion_index_dict
```

```python
def combination_preprocessing(string_format):
    idx_list = [0 for _ in range(len(combination_index_dict))]
    idx_dict = {}
    string_dict = json.loads(string_format)
    for key, value in string_dict.items():
        idx_list[combination_index_dict[key]] = value
        
    return idx_list


def champion_preprocessing(string_format):
    idx_list = [0 for _ in range(60)]
    string_dict = json.loads(string_format)
    for idx, (key, value) in enumerate(string_dict.items()):
        start_pos = idx * 5
        idx_list[start_pos] = champion_index_dict[key]
        for item_idx, item in enumerate(value["items"], start=(start_pos + 1)):
            idx_list[item_idx] = item
        idx_list[start_pos + 4] = value["star"]
        
    return idx_list
```

```python
df = pd.read_csv("TFT_Challenger_MatchData.csv")
df = df[["combination", "champion", "Ranked"]]

# # Get the length of combination and champion
# df["combination_length"] = df.combination.apply(len)
# df["champion_length"] = df.champion.apply(len)

# Check to dictionary
df.combination = df.combination.apply(combination_preprocessing)
df.champion = df.champion.apply(champion_preprocessing)

# Remove empty data
df.drop(df[(df.combination == dict()) & (df.champion == dict())].index, inplace=True)
df.drop(df[df.champion == dict()].index, inplace=True)


df.head()
```

```python
for combination, idx in combination_index_dict.items():
    df[f"combination_{combination}"] = df.combination.apply(lambda x: x[idx])
```

```python
for i in range(12):
    start_pos = i * 5
    df[f"character_{i + 1}_name"] = df.champion.apply(lambda x: x[start_pos])
    df[f"character_{i + 1}_item_1"] = df.champion.apply(lambda x: x[start_pos + 1])
    df[f"character_{i + 1}_item_2"] = df.champion.apply(lambda x: x[start_pos + 2])
    df[f"character_{i + 1}_item_3"] = df.champion.apply(lambda x: x[start_pos + 3])
    df[f"character_{i + 1}_star"] = df.champion.apply(lambda x: x[start_pos + 4])
```

```python
df.drop(columns=['combination', 'champion'], inplace=True)
```
